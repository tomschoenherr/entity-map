```python
import pandas as pd 
import os
import numpy as np
```

```python
from cleaning_fx import clean_df
```

```python
data_path = "data/"
```

```python
# Load the files for entity a
a_company = pd.read_csv(os.path.join(data_path, "a__company.csv"))
a_company.head(10)
```

Listing a_company irrelevant columns that I can drop:
- load_date, bucket_id, and lvl seem to be some sort of internal values
- address1 and address2 are redundant to address
- score doesn't have any discernible meaning
- cnt_children & cnt_opp doesn't seem to have any match to b_company features
- dunsnumber & parentdunsnumber 
- fax is all missing
- country is in geo info


```python

a_company['zipcode'] = a_company['zipcode'].replace(['0', 0], np.nan)
```

```python
a_company.head(5)
```

```python
a_geo = pd.read_csv(os.path.join(data_path, "a__geo.csv"))
a_geo.head(10)
```

features to drop from a_geo:
- is_primary, latitude, longitude, elevation, state_full_name, city_display, timezone, daylight_saving, region, division, congress_district, congress_land_area

```python
a_data = pd.merge(left = a_company,
                  right = a_geo,
                  on='geo_id',
                  how = 'left')

# Combine zipcodes with preference for zipcode_x, fallback to zipcode_y
a_data['zipcode'] = a_data['zipcode_x'].fillna(a_data['zipcode_y'])

a_data.rename(columns={"country_iso2": "iso", 'parent_vendor_id': 'parent_id', 'top_vendor_id': 'top_id', "vendor_id": "unique_id"}, inplace=True)

a_clean = clean_df(a_data)

a_clean.head(10)
```

```python
# Load the files for entity b
b_company = pd.read_csv(os.path.join(data_path, "b__company.csv"))
b_company.head(10)
```

Unnecessary features from b_company: 
- entity_proper_name, primary_sic_code, industry_code, sector_code, entity_type, entity_sub_type, year_founded, nace_code

- I already have the state info, so I want to drop the second part of the metro_area to at least keep the city name
- I should remove the special characters from each 

```python
b_hierarchy = pd.read_csv(os.path.join(data_path, "b__hierarchy.csv"))
b_hierarchy.head(10)
```

```python
# merge b_hierarchy and b_company by the b_entity_id
b_data = pd.merge(left = b_hierarchy,
                  right = b_company,
                  on='b_entity_id',
                  how = 'left')
```

```python
b_address = pd.read_csv(os.path.join(data_path, "b__address.csv"))
b_address.head(10)
```

What to drop from b_address:
- dataset a has no fax info, so drop fax, fax_country, fax_area, fax_full 

```python
b_address['address'] = (
    b_address['location_street1'].fillna('') + ' ' + 
    b_address['location_street2'].fillna('') + ' ' + 
    b_address['location_street3'].fillna('')
).str.strip()

b_address.head()
```

```python
b_data = pd.merge(left = b_data,
                  right = b_address,
                  on='b_entity_id',
                  how = 'left')

b_data.rename(columns = {"b_entity_id": "unique_id","entity_name": "name", 'web_site': 'websiteurl',
                            'b_parent_entity_id': 'parent_id', "tele_area": "area_code",
                           'b_ultimate_parent_entity_id': 'top_id', "tele_full": "phone", "location_city": "city"},
                             inplace=True)

b_data['iso'] = b_data['iso_country_x'].fillna(b_data['iso_country_y'])
b_data['state'] = b_data['state_province_x'].fillna(b_data['state_province_y'])
b_data['zipcode'] = b_data['zip_postal_code'].fillna(b_data['location_postal_code'])
```

```python
b_clean = clean_df(b_data)
b_clean.head(50)
```
